```python
employment_types = ["Full-time", "Part-time", "Contract", "Freelance"]
filtered_df = ds_salaries_selected_us_df[ds_salaries_selected_us_df['Employment Type'].isin(employment_types)]
plt.figure(figsize=(10, 5)) 
plt.bar(filtered_df['Employment Type'], filtered_df['Salary in USD'])
plt.title("Earning Potential by Employment Type in the U.S.", color = "red")
plt.xlabel("Employment Type", color = "red")
plt.ylabel("Salary in $USD", color = "red")
plt.grid(True)
plt.show()
```

```python
df_work = ds_salaries_selected_us_year_df.groupby(['Work Year', 'Employment Type'])['Salary in USD'].mean().unstack()
plt.figure(figsize=(20, 15))
df_work.plot()
plt.title("Average Salary by Employment Type in the U.S. by Year", color = "red")
plt.xlabel('Year', color = "red")
plt.ylabel('Avg. Salary in $USD', color = "red")
plt.xticks(rotation=60)
plt.legend(title='')
plt.tight_layout()
plt.show()
```

Question: What is the average salary by employment type in the U.S. by year?

Findings:
1. There is incomplete data in 2024 for contract and freelance employment types.
2. Contract had an unusual spike in 2021 with an average salary of $250K and showing a steady increase from approximately $70K to $100K for the period 2020 to 2023. 
3. Freelance shows an inconsistent trend ranging from $40K to $80K for the period 2020 to 2023.
4. Part-time shows a significant increase from $20K in 2020 to $80 in 2022 with a significant decrease in 2024 to $20K
5. Full-time shows a steady increase from $105K to $152K for the period 2020 to 2024.
